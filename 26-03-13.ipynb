```python
import numpy as np

def activation(func):
    def wrapper(x1, x2):
        return np.sign(func(x1, x2))
    return wrapper

@activation
def n11(x1, x2):
    return -x1+x2-1/2

@activation
def n12(x1, x2):
    return -x1+x2+1/2

@activation
def n21(x1, x2):
    return x1+x2-1

@activation
def n22(x1, x2):
    return -x1+x2-1/2

@activation
def n23(x1, x2):
    return -x1-x2-1/2

def neural_network(x1, x2):
    n11_out = n11(x1, x2)
    n12_out = n12(x1, x2)
    n21_out = n21(n11_out, n12_out)
    n22_out = n22(n11_out, n12_out)
    n23_out = n23(n11_out, n12_out)
    return n21_out, n22_out, n23_out

neural_network(0.8, 0.2)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

plt.scatter(X, Y, c=np.array(neural_network(X, Y)).argmax(axis=0))
plt.grid()
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```
